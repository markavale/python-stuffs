```python
from datasets import load_dataset
```

```python
data_name = "mlabonne/guanaco-llama2-1k"
training_data = load_dataset(data_name, split="train")
```

```python
training_data[150]
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import nltk
from collections import Counter
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
text = " ".join([item['text'] for item in training_data])
```

```python
# Tokenization (splitting the text into words)
words = nltk.word_tokenize(text)

# Removing punctuation and making all the words lowercase
words = [word.lower() for word in words if word.isalpha()]

# Removing stopwords from all available languages
all_stopwords = set()
for lang in stopwords.fileids():
    all_stopwords.update(stopwords.words(lang))

words = [word for word in words if word not in all_stopwords]

# Counting word frequencies
word_counter = Counter(words)

# Most common words
common_words = word_counter.most_common(10)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(*zip(*common_words))
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.title('Top 10 Most Common Words')
plt.show()
```
